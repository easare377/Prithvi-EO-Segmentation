```python
DEFAULT_ENCODING = {
    3: [255, 255,   0],  # yellow
    2: [180,  96,   0],  # brown-ish
    1: [251,  72, 196],  # magenta
    0: [  0,   0,   0],  # background
}
#function to covert multiclass mask to rgb and save as png
def save_mask_as_png(mask, path, encoding=DEFAULT_ENCODING):
    import numpy as np
    from PIL import Image

    h, w = mask.shape
    rgb_mask = np.zeros((h, w, 3), dtype=np.uint8)
    for class_id, color in encoding.items():
        rgb_mask[mask == class_id] = color
    img = Image.fromarray(rgb_mask)
    img.save(path)
```

```python
import image_utils as iu
import numpy as np
mask = iu.read_image(r'Y:\SPOT\2023\Asare\SSM_Planet\original\mask\Planet_21-0000023552-0000023552.png')
mask = np.array(mask)
out_path = 'Planet_21-0000023552-0000023552_rgb.png'   
save_mask_as_png(mask, out_path, encoding=DEFAULT_ENCODING)
```
